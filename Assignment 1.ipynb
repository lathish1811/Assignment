```python
"""Question 1(A) : Analyze Student Exam Scores with Python
Problem Statement:
You are provided with the exam scores of 20 students in a class. Write a Python program to analyze these scores. Your analysis should include:
1. Overall class performance – Calculate average, highest, and lowest scores.
2. Pass/fail classification – Identify which students passed the exam (assume pass mark is 50).
3. Grading on a curve – Assign letter grades (A, B, C, D, F) based on curved scoring or predefined thresholds."""
import numpy as np
import pandas as pd

scores = np.array([85, 92, 76, 48, 59, 67, 90, 73, 55, 34, 88, 62, 45, 100, 81, 79, 40, 70, 66, 53])
df = pd.DataFrame({'Score': scores})

#1. Calculate performance statistics
average_score = np.mean(scores)
highest_score = np.max(scores)
lowest_score = np.min(scores)
print("=== Overall Class Performance ===")
print(f"Average Score: {average_score:.2f}")
print(f"Highest Score: {highest_score}")
print(f"Lowest Score : {lowest_score}")

#2. Add Pass/Fail classification
def pass_fail(x):
    return 'Pass' if x >= 50 else 'Fail'
df['Pass/Fail'] = df['Score'].map(pass_fail)

#3. Add Grade column using predefined thresholds
def assign_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 50:
        return 'D'
    else:
        return 'F'
df['Grade'] = df['Score'].apply(assign_grade)
print("\n=== Detailed Student Analysis ===")
display(df)
```

```python
"""Question 1 (B) : Solve a System of Linear Equations
Problem Statement:
Write a Python program to solve a system of linear equations represented by:
                                Ax=b
Where:
 A is a 2x2 matrix (list of lists or NumPy array),
 b is a 2x1 vector,
 The program should compute and return the solution vector x."""

import numpy as np

# Define matrix A and vector b
A = np.array([[3, 1],[1, 2]])
b = np.array([9, 5])

# Solve for x using NumPy
x = np.linalg.solve(A, b)

# Display solution
print("=== Solution of Ax = b ===")
print(f"Matrix A:\n{A}")
print(f"Vector b:\n{b}")
print(f"Solution x:\n{x}")
```

```python
"""Question 2(A) : Analyze Daily Step Count Using NumPy
Problem Statement:
Write a Python program using NumPy to analyze your daily step count data collected over
the past month. The program should:
 Compute key statistics such as total steps, average steps per day, and
maximum/minimum step counts.
 Identify days with high activity (e.g., days with steps above a certain threshold).
 Analyze trends over time, such as increases or decreases in activity, and optionally
visualize them."""
import numpy as np
import matplotlib.pyplot as plt

# Generate or input your daily step count data (for 30 days)
# For this example, we'll generate random data within a reasonable range
np.random.seed(42)  # Set a seed for reproducibility
step_counts = np.random.randint(3000, 12000, 30)  # Random daily steps between 3000 and 12000

# Compute key statistics
total_steps = np.sum(step_counts)     # Total steps
average_steps = np.mean(step_counts)  # Average steps per day
max_steps = np.max(step_counts)       # Maximum steps in a day
min_steps = np.min(step_counts)       # Minimum steps in a day

# Display statistics
print("Key Statistics:")
print(f"Total Steps: {total_steps}")
print(f"Average Steps per Day: {average_steps:.2f}")
print(f"Maximum Steps in a Day: {max_steps}")
print(f"Minimum Steps in a Day: {min_steps}")

# Identify high-activity days (e.g., steps above 10,000)
threshold = 10000  #Define the threshold for high activity
high_activity_days = np.where(step_counts > threshold)[0]  # Find indices where steps exceed threshold

# Display high-activity days
print(f"\nHigh Activity Days (Steps above {threshold}): {high_activity_days + 1}")  # +1 to make days 1-indexed

# Analyze trends over time (simple increase or decrease check)
daily_diff = np.diff(step_counts)  # Calculate daily difference in step counts (increase/decrease)
increase_days = np.where(daily_diff > 0)[0]  # Days where activity increased
decrease_days = np.where(daily_diff < 0)[0]  # Days where activity decreased

# Display trend analysis
print(f"\nDays with increased activity: {increase_days + 1}")  # +1 for 1-indexing
print(f"Days with decreased activity: {decrease_days + 1}")  # +1 for 1-indexing

# Plot daily step count data
plt.figure(figsize=(10, 6))
plt.plot(step_counts, label='Daily Step Count', marker='o', color='b')
plt.axhline(y=threshold, color='r', linestyle='--', label=f'High Activity Threshold ({threshold} steps)')
plt.title('Daily Step Count Over the Month')
plt.xlabel('Day of the Month')
plt.ylabel('Steps')
plt.legend()
plt.grid(True)
plt.show()
```

```python
"""Question 2 (B) : Analyze Student Grades Using NumPy
Problem Statement:
Write a Python program using NumPy to analyze the grades of 30 students across three
subjects: Math, Science, and English. Your program should:
 Calculate overall and subject-wise statistics such as average, highest, and lowest
grades.
 Identify top-performing students based on total or average scores.
 Optionally visualize the grade distributions for each subject using histograms or bar
charts."""
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate random grades for 30 students in 3 subjects (Math, Science, English)
np.random.seed(42)  # For reproducibility
grades = np.random.randint(0, 101, size=(30, 3))  # Random grades between 0 and 100

# Step 2: Calculate overall and subject-wise statistics
# Overall statistics
total_scores = grades.sum(axis=1)  # Total score for each student
average_scores = grades.mean(axis=1)  # Average score for each student

# Subject-wise statistics
subject_averages = grades.mean(axis=0)  # Average for each subject
subject_highest = grades.max(axis=0)  # Highest score in each subject
subject_lowest = grades.min(axis=0)  # Lowest score in each subject

# Overall highest and lowest scores
overall_highest = grades.max()
overall_lowest = grades.min()

# Identify top-performing students (based on total score or average score)
top_student_index = np.argmax(total_scores)  # Student with the highest total score
top_student_avg_index = np.argmax(average_scores)  # Student with the highest average score

# Displaying results
print("Subject-wise Statistics:")
print(f"Math: Avg = {subject_averages[0]:.2f}, Highest = {subject_highest[0]}, Lowest = {subject_lowest[0]}")
print(f"Science: Avg = {subject_averages[1]:.2f}, Highest = {subject_highest[1]}, Lowest = {subject_lowest[1]}")
print(f"English: Avg = {subject_averages[2]:.2f}, Highest = {subject_highest[2]}, Lowest = {subject_lowest[2]}")
print(f"\nOverall Statistics:")
print(f"Overall Highest Score: {overall_highest}")
print(f"Overall Lowest Score: {overall_lowest}")
print(f"\nTop-performing Student (based on total score): Student #{top_student_index + 1}")
print(f"Top-performing Student (based on average score): Student #{top_student_avg_index + 1}")

# Step 3: Visualizing the grade distributions for each subject

subjects = ['Math', 'Science', 'English']

# Create histograms for each subject
plt.figure(figsize=(10, 6))

for i in range(3):  # Loop through the 3 subjects
    plt.subplot(1, 3, i+1)  # Create 1x3 grid of subplots
    plt.hist(grades[:, i], bins=10, color='skyblue', edgecolor='black')
    plt.title(f'{subjects[i]} Grade Distribution')
    plt.xlabel('Grade')
    plt.ylabel('Frequency')
    
plt.tight_layout()
plt.show()
```

```python
"""Question 3(A) : Compute Total Impedance of an Electrical Circuit
Problem Statement:
Write a Python program to calculate the total impedance of an electrical circuit using the
given impedance values of individual components—resistors, capacitors, and inductors.
Your program should be able to:
 Compute total impedance for components connected in series and parallel.
 Support impedance values in complex form (to account for frequency-dependent
elements like capacitors and inductors).
 Optionally allow the user to specify the circuit configuration and component values."""
    
import numpy as np

# Function to calculate impedance of inductor: Z = jωL
def impedance_inductor(L, freq):
    return 1j * 2 * np.pi * freq * L

# Function to calculate impedance of capacitor: Z = 1 / (jωC)
def impedance_capacitor(C, freq):
    return 1 / (1j * 2 * np.pi * freq * C)

# Function to calculate total series impedance
def total_impedance_series(components):
    return sum(components)

# Function to calculate total parallel impedance
def total_impedance_parallel(components):
    inv_sum = sum(1 / z for z in components)
    return 1 / inv_sum

# Example usage
frequency = 50  # Hz

# Component values
R = 100  # Ohm
L = 0.1  # Henry
C = 100e-6  # Farad

# Convert components to complex impedance
Z_R = R
Z_L = impedance_inductor(L, frequency)
Z_C = impedance_capacitor(C, frequency)

# Display individual impedances
print(f"Resistor Impedance (Z_R): {Z_R} Ω")
print(f"Inductor Impedance (Z_L): {Z_L:.2f} Ω")
print(f"Capacitor Impedance (Z_C): {Z_C:.2f} Ω")

# Choose configuration
configuration = 'series'  # 'series' or 'parallel'

if configuration == 'series':
    Z_total = total_impedance_series([Z_R, Z_L, Z_C])
elif configuration == 'parallel':
    Z_total = total_impedance_parallel([Z_R, Z_L, Z_C])
else:
    Z_total = None
    print("Invalid configuration selected.")

# Output result
if Z_total is not None:
    print(f"\nTotal {configuration.capitalize()} Impedance: {Z_total:.2f} Ω")
```

```python
"""Question4 (B): Perform Basic Data Analysis on Product Inventory Using Pandas
Problem Statement:
You are given a dictionary containing information about products in an inventory, including
their names, prices, and quantities available:

data = {'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],'Price': [1200, 800, 300, 150, 50],'Quantity': [10, 20, 15, 25, 50]}"""

import pandas as pd
import matplotlib.pyplot as plt

# Product inventory dictionary
data = {'Product': ['Laptop', 'Smartphone', 'Tablet', 'Monitor', 'Keyboard'],'Price': [1200, 800, 300, 150, 50],'Quantity': [10, 20, 15, 25, 50]}

# Create DataFrame
df = pd.DataFrame(data)

# 1. Total inventory value per product
df['Total Value'] = df['Price'] * df['Quantity']

# 2. Find most expensive product
most_expensive = df.loc[df['Price'].idxmax()]

# 3. Find product with highest quantity
most_stocked = df.loc[df['Quantity'].idxmax()]

# 4. Calculate average price and quantity
avg_price = df['Price'].mean()
avg_quantity = df['Quantity'].mean()

# Display outputs
print("Full Inventory Data:\n", df)
print("\nMost Expensive Product:\n", most_expensive)
print("\nProduct with Highest Quantity:\n", most_stocked)
print("\nAverage Price:", avg_price)
print("Average Quantity:", avg_quantity)

# Plot: Bar chart of Total Value per Product
plt.figure(figsize=(8, 5))
plt.bar(df['Product'], df['Total Value'], color='skyblue')
plt.title('Total Inventory Value per Product')
plt.xlabel('Product')
plt.ylabel('Total Value ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
"""Question5 (A): Perform Sales Data Analysis Using Pandas
Problem Statement:
You are given a CSV file named sales_data.csv containing sales records for a company.
Write a Python script using Pandas to perform the following tasks:
1. Read the CSV file into a Pandas DataFrame.
2. Display the first 5 rows of the DataFrame to get an overview of the data.
3. Check for missing values and handle them appropriately (e.g., fill or drop).
4. Visualize the monthly sales trend using a line plot to better understand sales
performance over time.
You may use Matplotlib or Seaborn for visualization."""

import pandas as pd
import matplotlib.pyplot as plt

# 1. Read the Excel file
df = pd.read_excel('sales.xlsx')  # Replace with your actual filename
print("Original Dataframe:\n",df)

# 2. Display the first 5 rows
print("First 5 rows of the data:")
print(df.head())

# 3. Clean the 'Value' column
df['Value'] = df['Value'].replace(r'[\$,]', '', regex=True)
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')

# Show missing value count
print("\nMissing values before filling:")
print(df['Value'].isnull().sum())

# 4. Fill missing values using forward fill (pad)
df['Value'] = df['Value'].ffill().astype(int)
print("DataFrame after filling missing values in 'Value':")
print(df)

# Show missing value count after filling
print("\nMissing values after filling:")
print(df['Value'].isnull().sum())

# 5. Group by year and sum the sales
yearly_sales = df.groupby('Year')['Value'].sum().reset_index()
print(yearly_sales)

# 6. Plot the yearly trend
plt.figure(figsize=(8, 5))
plt.plot(yearly_sales['Year'], yearly_sales['Value'], marker='o', linestyle='-', color='green')
plt.title('Yearly Sales Trend (Forward Filled)')
plt.xlabel('Year')
plt.ylabel('Total Sales Value')
plt.grid(True)
plt.tight_layout()
plt.show()

```
